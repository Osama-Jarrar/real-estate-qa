```python
# ✅ Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ✅ Load dataset
file_path = "../Data/kc_house_data.csv"  # adjust as needed
df = pd.read_csv(file_path)

```

```python
# ✅ Basic info
print("=== Shape ===")
print(df.shape)

print("\n=== Columns and Types ===")
print(df.dtypes)

print("\n=== Sample ===")
display(df.head(5))

```

```python
# Show all rows that have duplicated IDs
df[df.duplicated(subset="id", keep=False)].sort_values("id")
```

```python
df.yr_renovated
```

```python
# ✅ Missing values
print("\n=== Missing Values ===")
print(df.isnull().sum())

```

```python
# ✅Basic visualizations
sns.histplot(df["price"], kde=True)
plt.title("Price Distribution")
plt.show()

sns.boxplot(x="bedrooms", y="price", data=df)
plt.title("Price vs Bedrooms")
plt.show()
```
